```python
from ultralytics import YOLO
import cv2
import torch
import matplotlib.pyplot as plt
from IPython.display import Image, display
```

```python
model = YOLO("yolov8m.pt")
```

```python
def detect_person_yolov8(image_name):
    # The code below this comment will execute the YOLOv8 model on the image
    model = YOLO("yolov8m.pt")
    results = model.predict(image_name)

    # Load the image using OpenCV
    image = cv2.imread(image_name)
    image_with_boxes = image.copy()  # Create a copy of the image to draw bounding boxes on

    # Iterate through the detected objects and draw bounding boxes
    for box in results[0].boxes:
        
        class_name = results[0].names[box.cls[0].item()]
        if class_name == "person":
            confidence = round(box.conf[0].item(), 2)
            cords = box.xyxy[0].tolist()
            cords = [round(x) for x in cords]

            # Extract coordinates
            x1, y1, x2, y2 = cords

            # Draw bounding box on the image
            cv2.rectangle(image_with_boxes, (x1, y1), (x2, y2), (0, 255, 0), 2)

            # Add class label and confidence score to the image
            label = f"{class_name}: {confidence}"
            cv2.putText(image_with_boxes, label, (x1 + 5, y1 + 15), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

            # Print information
            print("Object type:", class_name)
            print("Probability:", confidence)
        
    # Display the image with bounding boxes and confidence scores
    # cv2.imshow("Image with Bounding Boxes", image_with_boxes)

    output_image_path = "output_image.jpg"
    cv2.imwrite(output_image_path, image_with_boxes)
    display(Image(filename=output_image_path))
#     cv2.waitKey(0)
#     cv2.destroyAllWindows()
```
